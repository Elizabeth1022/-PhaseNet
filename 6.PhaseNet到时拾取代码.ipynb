    kpsh:即使峰值比“mpd”更接近，也要保持相同的高度;<br>
    vally：波峰or波谷;<br>
    show:是否显示;<br>
    ax:坐标轴长度;<br>

* numpy isfinite（）：按元素进行测试，无论它是否有限（不是无穷大，不是数字），然后将结果作为布尔数组返回。<br>
  * 参量：infinite（）函数采用两个参数，其中一个是可选参数，而一个则是必需参数。<br>
    第一个参数是输入数组或我们要检查其是否有限的输入。<br>
    第二个是n维数组，它是可选的。将输出数组与结果放置在一起。<br>
  * 返回值：infinite（）函数返回一个布尔数组，如果我们传递一个数组，则返回结果；如果传递标量值，则返回布尔值true或false。<br>
